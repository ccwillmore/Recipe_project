```python
corrected_ingredient_list = []
for index, element in enumerate(ingredient_list):
    if re.search('(?:^|\s)[Cc]hicken(?!sstock|\sbroth|\sbouillon|\sbase)',element):
       corrected_ingredient_list.append('chicken')
    elif re.search('(?:$|\s)[Pp]ork|[Bb]aby\sback|[Bb]oston\sbutt(?:\s|$)',element):
       corrected_ingredient_list.append('pork')
    elif re.search('(?:$|\s)bean(?!\ssauce)',element):
       corrected_ingredient_list.append('beans') 
    elif re.search('(?:(?<=acorn)|(?<=butternut))\s*squash(?:$|\s)', element):
       corrected_ingredient_list.append('winter squash')
    elif re.search('(?:(?<=american\s)|(?<=cottage\s)|(?<=mexican\s)|(?<=cheddar\s)|(?<=shredded\s)|(?<=swiss\s)|(?<=cream\s)|(?<=^))cheese(?:$|\s)', element):
       corrected_ingredient_list.append('cheese')
    elif re.search('(?:(?<=asiago\s)|(?<=ricotta\s)|(?<=mozzarella\s)|(?<=parmessan\s)|(?<=^))cheese(?:$|\s)', element):
       corrected_ingredient_list.append('italian cheese')
    elif re.search('(?:(?<=black\s)|(?<=green\s)|(?<=pink\s))[Pp]epper(?:corns|\s|$)', element):
       corrected_ingredient_list.append('pepper')
    elif re.search('(?:(?<=blue\s)|(?<=feta\s)|(?<=goat\s)|(?<=^))cheese(?:$|\s)', element):
       corrected_ingredient_list.append('mediterranean cheese')
    elif re.search('(?:(?<=green)|(?<=yellow)|(?<=orange)|(?<=red)|(?<=bell))\spepper(?:$|\s)', element):
       corrected_ingredient_list.append('bell pepper')
    elif re.search('(?:^|\s)[Aa]maranth|[Bb]arley|[Qq]uinoa(?:$|\s)',element):
       corrected_ingredient_list.append('grain')
    elif re.search('(?:^|\s)[Aa]ncho(?:$|\s)', element):
       corrected_ingredient_list.append('ancho pepper')
    elif re.search('(?:^|\s)[Aa]nchovy(?:$|\s)', element):
       corrected_ingredient_list.append('anchovy')
    elif re.search('(?:^|\s)[Aa]pple(?:$)', element):
       corrected_ingredient_list.append('apple')
    elif re.search('(?:^|\s)[Aa]pple\s(?:cider|juice)(?!\svinegar)', element):
       corrected_ingredient_list.append('apple juice')
    elif re.search('(?:^|\s)[Aa]pricot(?:$|\s)', element):
       corrected_ingredient_list.append('apricot')
    elif re.search('(?:^|\s)[Bb]alsamic(?:\sglaze|$)', element):
       corrected_ingredient_list.append('balsamic')
    elif re.search('(?:^|\s)[Bb]acon(?:$|\s)', element):
       corrected_ingredient_list.append('bacon')
    elif re.search('(?:^|\s)[Bb]anana\spepper(?:$|\s|s)', element):
       corrected_ingredient_list.append('fresh pepper')
    elif re.search('(?:^|\s)[Bb]ase|[Ss]tock|[Bb]ouillon|[Bb]roth(?:$|\s)', element):
       corrected_ingredient_list.append('broth')
    elif re.search('(?:^|\s)[Bb]asil(?!\spesto)', element):
       corrected_ingredient_list.append('basil')
    elif re.search('(?:^|\s)[Bb]eef(?!\sbroth|\sbase|\sstock|\sbouillon|steak|\sgravy)', element):
       corrected_ingredient_list.append('beef')
    elif re.search('(?:^|\s)[Bb]eet(?:$|\s|s)', element):
       corrected_ingredient_list.append('beets')
    elif re.search('(?:(?:^)|(?:\s))[Cc]arrots(?:$|\s)', element):
       corrected_ingredient_list.append('carrots')
    elif re.search('(?:^|\s)[Cc]ardamom(?:$|\s|s)', element):
       corrected_ingredient_list.append('cardomom')
    elif re.search('(?:^|\s)[Cc]hocolate(?:$|\s)',element):
       corrected_ingredient_list.append('chocolate')
    elif re.search('(?:^|\s)[cC]orn(?:$|\s)', element):
       corrected_ingredient_list.append('corn')
    elif re.search('(?:^|\s)[Ee]ggplant(?:$|\s|s)', element):
       corrected_ingredient_list.append('eggplant')
    elif re.search('(?:^|\s)[Gg]ravy(?:$|\s)', element):
       corrected_ingredient_list.append('gravy')
    elif re.search('(?:^|\s)[Ll]entil(?:$|\s|s)', element):
       corrected_ingredient_list.append('lentils')
    elif re.search('(?:^|\s)[Mm]ilk(?:$|\s)',element):
       corrected_ingredient_list.append('milk')
    elif re.search('(?:^|\s)[Oo]live(?!\soil)', element):
       corrected_ingredient_list.append('olives')
    elif re.search('(?:^|\s)[Pp]asta(?:$|\s)', element):
       corrected_ingredient_list.append('pasta')
    elif re.search('(?:^|\s)[Pp]ear(?:$|\s|s)', element):
       corrected_ingredient_list.append('pear')
    elif re.search('(?:^|\s)[Pp]esto(?:$|\s)', element):
       corrected_ingredient_list.append('pesto')
    elif re.search('(?:^|\s)[Rr]ice(?!\sflour|\spudding|d)(?:$|\s)', element):
       corrected_ingredient_list.append('rice')
    elif re.search('(?:^|\s)[Ss]pinach(?:$|\s|s)', element):
       corrected_ingredient_list.append('spinach')
    elif re.search('(?:^|\s)[Tt]una(?:$|\s)',element):
       corrected_ingredient_list.append('fish')
    elif re.search('(?:^|\s)[Tt]omato\w*(?!\spaste|\ssauce)', element):
       corrected_ingredient_list.append('tomato')
    elif re.search('(?:^|\s)[Vv]inegar(?:$|\s)', element):
       corrected_ingredient_list.append('vinegar')
    elif re.search('(?:^|\s)[Vv]anilla(?!\sice|\scake)', element):
       corrected_ingredient_list.append('vanilla')
    elif re.search('(?:^|\s)masala(?:$|\s)', element):
       corrected_ingredient_list.append('masala')
    elif re.search('(?:^|\s|\w)berries|berry(?:$|\s)', element):
       corrected_ingredient_list.append('berries')
    elif re.search('(?:black|green|pink|\s)[Pp]epper(?:corns|\s|$)', element):
       corrected_ingredient_list.append('black pepper')
    elif re.search('(?:^|\s)[Aa]lmond(?!\smilk)', element):
       corrected_ingredient_list.append('almonds')
    elif re.search('(?:^|\s)[Oo]il(?:\s|$)', element):
       corrected_ingredient_list.append('oil')
    elif re.search('(?:^|\s)[Aa]vocado(?!\soil)', element):
       corrected_ingredient_list.append('avocado')
    elif re.search('(?:^|\s)[Tt]urkey(?!\sstock|\sbroth)', element):
       corrected_ingredient_list.append('turkey')
    elif re.search('(?:^|\s)[Bb]read(?!\sflour|crumbs)', element):
       corrected_ingredient_list.append('bread')
    elif re.search('(?:^|\s)[Bb]ratwurst|[Cc]horizo|[Ss]ausage(?:$|\s)', element):
       corrected_ingredient_list.append('sausage')
    elif re.search('(?:^|\s)[Bb]roccoli(?:$|\sfloret)', element):
       corrected_ingredient_list.append('broccoli')
    elif re.search('(?:^|\s)[Mm]ustard\sseed(?:$|s)', element):
       corrected_ingredient_list.append('mustard seed')
    elif re.search('(?:(?:^)|(?:clarified\s)|(?:unsalted\s))[Bb]utter(?!\slettuce)', element):
       corrected_ingredient_list.append('butter')
    elif re.search('(?:^|\s)[Cc]]abbage(?:$|/s)', element):
       corrected_ingredient_list.append('cabbage')
    elif re.search('(?:^|\s)[Cc]urry\spaste(?:$|/s)', element):
       corrected_ingredient_list.append('curry paste')
    elif re.search('(?:^|\s)[Pp]ineapple(?:$|/s)', element):
       corrected_ingredient_list.append('pineapple')
    elif re.search('(?:(?:himalayan\s)|(?:coarse\s)|(?:sea\s)|($))[Ss]alt(?:$|/s)', element):
       corrected_ingredient_list.append('salt')
    elif re.search('(?:(?:\s)|(?:$))[Mm]ozzarella(?:$|/s)', element):
       corrected_ingredient_list.append('italian cheese')
    elif re.search('(?:^|\s)[Bb]utterscotch(?:$|/s)', element):
       corrected_ingredient_list.append('butterscotch')
    elif re.search('(?:(?:^)|(?:bread\s)|(?:cake\s)|wheat\s)[Ff]lour(?:$|/s)', element):
       corrected_ingredient_list.append('wheat flour')
    elif re.search('(?:(?:^)|(?:candied\s)|(?:crystalized\s)|(?:powdered\s)|(?:ground\s))[Gg]inger(?:$|/s)', element):
       corrected_ingredient_list.append('ginger')
    elif re.search('(?:(?:^)|(?:\s))[Bb]lackeyed\speas(?:$|/s)', element):
       corrected_ingredient_list.append('blackeyed peas')
    elif re.search('(?:(?:^)|(?:\s))[Oo]nion(?!\ssoup)', element):
       corrected_ingredient_list.append('onion')
    elif re.search('(?:(?:green\s)|(?:chipotle\s))[Cc]hile(?:\s|s|$)', element):
       corrected_ingredient_list.append('fresh chile')
    elif re.search('(?:(?:^)|(?:\s))[Cc]auliflower(?:\sfloret|$)', element):
       corrected_ingredient_list.append('cauliflower')
    elif re.search('(?:(?:^)|(?:\s))[Cc]elery(?!\ssalt|\sseed)', element):
       corrected_ingredient_list.append('celery')
    elif re.search('(?:(?:^)|(?:\s))[Cc]herry(?:\s|$)', element):
       corrected_ingredient_list.append('cherry')
    elif re.search('(?:(?:chickpea\s)|(?:rice\s)|(?:almond\s)|(?:coconut\s))[Ff]lour(?:\s|$)', element):
       corrected_ingredient_list.append('alt flour')
    elif re.search('(?:(?:^)|(?:\s))[Cc]innamon(?:\sstick|$)', element):
       corrected_ingredient_list.append('cinnamon')
     elif re.search('(?:(?:^)|(?:\s))[Cc]innamon(?:\sstick|$)', element):
       corrected_ingredient_list.append('cinnamon')
    else:
        corrected_ingredient_list.append(element)
corrected_ingredient_list
```

```python
corr_ing_list_df = pd.DataFrame(corrected_ingredient_list)
```

```python
corr_ing_list_df.rename(columns={0:'ingredient'}, inplace=True)
corr_ing_list_df['count']=1
```

```python
corr_ing_list_df = corr_ing_list_df.groupby('ingredient').agg({'count':'count'}).sort_values(by='count', ascending=False)
```
