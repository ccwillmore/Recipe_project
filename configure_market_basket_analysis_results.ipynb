```python
import pandas as pd
import numpy as np
```

```python
basket_analysis_spices_df = pd.read_csv('Exported_datafiles/Mexican_market_basket_analysis_results.csv', index_col = 0)
basket_analysis_spices_df
```

```python
basket_analysis_spices_df = basket_analysis_spices_df.rename(columns = {'0': 'entry_ing',
                                                                       '1':  'first_linked_ing',
                                                                       '2':  'second_linked_ing',
                                                                       '3':  'third_linked_ing',
                                                                       '4':  'fourth_linked_ing'})
```

```python
basket_analysis_spices_df
```

```python
matrix_rows = basket_analysis_spices_df.entry_ing.unique()
```

```python
matrix_rows
```

```python
first_link_rows = basket_analysis_spices_df.first_linked_ing.unique()
first_link_rows
```

```python
second_link_rows = basket_analysis_spices_df.second_linked_ing.unique()
second_link_rows
```

```python
third_link_rows = basket_analysis_spices_df.third_linked_ing.unique()
fourth_link_rows = basket_analysis_spices_df.fourth_linked_ing.unique()
fourth_link_rows
```

```python
matrix_columns = []
for element in first_link_rows:
    if (element not in matrix_columns and element is not np.nan):
        matrix_columns.append(element)
for element in second_link_rows:
    if (element not in matrix_columns and element is not np.nan):
        matrix_columns.append(element)
for element in third_link_rows:
    if element not in matrix_columns and element is not np.nan:
        matrix_columns.append(element)        
for element in third_link_rows:
    if element not in matrix_columns and element is not np.nan:
        matrix_columns.append(element)        

matrix_columns
```

```python
basket_analysis_spices_df.set_index('entry_ing')
```

```python
output_list = []

for row in range (0, len(matrix_rows)):
    result_row = []
    for col in range (0, len(matrix_columns)):
        result_row.append(0)
    output_list.append(result_row)
output_list
result_df = pd.DataFrame(output_list, columns = matrix_columns, index=matrix_rows)
result_df
```
