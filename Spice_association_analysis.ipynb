```python
spices_no_cuisine = spices_df.drop(columns='cuisine_SP').copy()
ingredient_names = spices_no_cuisine.columns.tolist()
```

```python
for element in ingredient_names:
    spices_no_cuisine.loc[:,element].replace({1:  element}, inplace=True)
spices_no_cuisine
```

```python
ingredient_list_of_lists = []

for row in spices_no_cuisine.index.values:
    rows_list = []
    rows_list = spices_no_cuisine.loc[row, :].tolist()
    ingredient_list_of_lists.append(rows_list)
```

```python
for index, row in enumerate(ingredient_list_of_lists):
    ingredient_list_of_lists[index] = [ing for ing in row if ing !=0]
    
```

```python
ingredient_list_of_lists = [ing for ing in ingredient_list_of_lists if len(ing) > 0]
```

```python
len(ingredient_list_of_lists)
```

```python
associations = apriori(ingredient_list_of_lists,  min_support = 0.005, min_confidence = 0.4, min_lift = 2)
```

```python
associations = list(associations)
```

```python
len(associations)
```

```python
with open ('Apriori_output', 'w') as f:
    for i in range (0,len(associations)):
        print(associations[i], file=f)
f.close()
```

```python
apriori_results_list = []
input_file = open('Apriori_output', 'r') 
apriori_results_list = input_file.readlines()
```

```python
final_groups = []

for i, item in enumerate(apriori_results_list):
    row_group = []
    row_group.append(re.search("(?<={)(.*?)(?=})",item).group())
#     row_group[0].remove("'")
    row_group = row_group[0].split(',')
    final_groups.append(row_group)
# final_groups
final_groups_df = pd.DataFrame(final_groups)
```

```python
for row in range(0, len(final_groups_df)):
    for index in final_groups_df.columns.tolist():
        temp_holder = final_groups_df.loc[row,index]
        try:
            final_groups_df.loc[row,index] = re.search("(?<=\')(.*?)(?=\')", temp_holder).group()
        except: continue
final_groups_df.replace({'None':  ''}, inplace=True)
```
