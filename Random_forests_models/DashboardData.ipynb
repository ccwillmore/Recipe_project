```python
import pandas as pd
```

```python
recipe_df = pd.read_csv("recipe_df_notencoded.csv")
```

```python
recipe_df
```

```python
empty_recipe_df = pd.read_csv("empty_recipe_df.csv")
```

```python
empty_recipe_df
```

```python
empty_recipe_df.rename(columns={"cuisine_predict": "cuisine_SP"})
```

```python
new_df=pd.concat([recipe_df,empty_recipe_df], axis=0, ignore_index=True)
```

```python
new_df.loc[new_df['cuisine_SP'].isnull(),'cuisine_SP'] = new_df['cuisine_predict']
```

```python
new_df
```

```python
new_df.drop(["cuisine_predict"], axis=1)
```

```python
total_dashboard = new_df.groupby(['cuisine_SP']).sum()
```

```python
total_dashboard.drop(["Unnamed: 0"], axis=1, inplace=True)
```

```python
total_dashboard
```

```python
new_df.to_csv("all_cuisines.csv")
```
