```python
# Initial imports.
import pandas as pd
import numpy as np
from path import Path
from ast import literal_eval
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# Reading the clean data
file = '../100_random_cleaned_0908.csv'
random_recipe_df = pd.read_csv(file, converters={"cuisine_SP":literal_eval})
random_recipe_df = random_recipe_df.drop(columns="Unnamed: 0")
random_recipe_df
```

```python
cuisine_name = []

for index in range(0, len(random_recipe_df)):
    if "Italian" in random_recipe_df.cuisine_SP[index]:
        cuisine_name.append("Italian")

    elif "Mexican" in random_recipe_df.cuisine_SP[index]:
        cuisine_name.append("Mexican")
    
    elif "American" in random_recipe_df.cuisine_SP[index]:
        cuisine_name.append("American")

    elif "Indian" in random_recipe_df.cuisine_SP[index]:
        cuisine_name.append("Indian")

    elif "Southern" in random_recipe_df.cuisine_SP[index]:
        cuisine_name.append("Southern")
    
    elif "Asian" in random_recipe_df.cuisine_SP[index]:
        cuisine_name.append("Asian")

    elif "English" in random_recipe_df.cuisine_SP[index]:
        cuisine_name.append("English")

    elif "Vietnamese" in random_recipe_df.cuisine_SP[index]:
        cuisine_name.append("Vietnamese")

    elif "French" in random_recipe_df.cuisine_SP[index]:
        cuisine_name.append("French")

    elif "Middle Eastern" in random_recipe_df.cuisine_SP[index]:
        cuisine_name.append("Middle Eastern")

    elif "Creole" in random_recipe_df.cuisine_SP[index]:
        cuisine_name.append("Creole")

    elif "Greek" in random_recipe_df.cuisine_SP[index]:
        cuisine_name.append("Greek")

    else:
        cuisine_name.append("")
     

print(cuisine_name)
```

```python
# Viewing data type for cusisine_sp
random_recipe_df.dtypes
```

```python
random_recipe_df.cuisine_SP = cuisine_name
```

```python
print(random_recipe_df['cuisine_SP'].value_counts())
```

```python
# Grouping all empty cuisines in it's dataframe and rearrange the index

Empty_recipe_df = random_recipe_df[random_recipe_df['cuisine_SP'] == ""].reset_index()

Empty_recipe_df.drop(columns='index', inplace=True, axis=1)

Empty_recipe_df
```

```python
# Grouping all recipes wiht a cuisine type and rearrange the index
recipe_df = random_recipe_df[random_recipe_df['cuisine_SP'] != ""].reset_index()
recipe_df.drop(columns='index', inplace=True, axis=1)
recipe_df
```

```python
#Define the features set
X = recipe_df.copy()
X = X.drop("cuisine_SP", axis=1)
X.head()
```

```python
# define the target

y = pd.get_dummies(recipe_df['cuisine_SP'])
y_italian = y['Italian']
y_italian
```

```python
# Splitting into Train and Test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y_italian, random_state=78)
```

```python
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=100, random_state=78) 
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data.
predictions = rf_model.predict(X_test_scaled)
predictions
```

```python
# Calculating the confusion matrix.
cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
italian_cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])
    
italian_cm_df
```
