```python
# Grouping all empty cuisines in it's dataframe and rearrange the index

Empty_recipe_df = random_recipe_df[random_recipe_df['cuisine_SP'] == ""]

#Empty_recipe_df.drop(columns='index', inplace=True, axis=1)

Empty_recipe_df
```

```python
# Grouping all recipes wiht a cuisine type and rearrange the index
recipe_df = random_recipe_df[random_recipe_df['cuisine_SP'] != ""]
#recipe_df.drop(columns='index', inplace=True, axis=1)
recipe_df
```

```python

le = LabelEncoder()
le.fit(recipe_df['cuisine_SP'])
le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(le_name_mapping)
```

```python
recipe_df['cuisine_SP'] = le.transform(recipe_df['cuisine_SP'])
recipe_df
```

```python
#Define the features set
X = recipe_df.copy()
X = X.drop("cuisine_SP", axis=1)
X.head()
```

```python
# define the target

y = recipe_df["cuisine_SP"]
y
```

```python
# Splitting into Train and Test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=100, random_state=78) 
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train, y_train)
```

```python
# Making predictions using the testing data.
predictions = rf_model.predict(X_test)
predictions
```

```python
# Calculating the confusion matrix.
cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
italian_cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1","Actual 2","Actual 3","Actual 4","Actual 5","Actual 6","Actual 7","Actual 8","Actual 9","Actual 10","Actual 11","Actual 12","Actual 13","Actual 14"], 
    columns=["Predicted 0", "Predicted 1","Predicted 2","Predicted 3","Predicted 4","Predicted 5","Predicted 6","Predicted 7","Predicted 8","Predicted 9","Predicted 10","Predicted 11","Predicted 12","Predicted 13","Predicted 14"])
    
italian_cm_df
```
