```python
import pandas as pd
import numpy as np
from apyori import apriori
from ast import literal_eval
```

```python
mexican_df = pd.read_csv('Mexican_cleaned_0908.csv')
mexican_df
```

```python
mexican_df.drop(columns=['Unnamed: 0', 'cuisine_SP'], inplace=True)
```

```python
mexican_df
```

```python
ingredient_list_of_lists = []
indexes = mexican_df.index.values
col_names = mexican_df.columns
for index in indexes:
    ingredient_list = []
    for ingredient in col_names:
        if mexican_df.loc[index,ingredient] == 1:
            mexican_df.loc[index,ingredient] = ingredient
        ingredient_list.append(mexican_df.loc[index,ingredient])
    ingredient_list_of_lists.append(ingredient_list)
ingredient_list_of_lists
```

```python
for index in range(0, len(ingredient_list_of_lists)):
    result = [ing for ing in ingredient_list_of_lists[index] if ing != 0]
    ingredient_list_of_lists[index] = result
ingredient_list_of_lists
            
```

```python
common_ingredients = ["salt and pepper",'water','lettuce', "oil", "tortilla",'onion','cheese','chicken','pork','beef','fish', 'beans']
```

```python
for index in range(0,len(ingredient_list_of_lists)):
    result = [ing for ing in ingredient_list_of_lists[index] if ing not in common_ingredients]
    ingredient_list_of_lists[index] = result
print(ingredient_list_of_lists)
```

```python
associations = apriori(ingredient_list_of_lists, min_length = 5, min_support = 0.075, min_confidence = 0.2, min_lift = 3)
```

```python
associations = list(associations)
```
